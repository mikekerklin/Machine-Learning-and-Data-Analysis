```python
import csv
import requests

url = 'https://vincentarelbundock.github.io/Rdatasets/csv/carData/TitanicSurvival.csv'
r = requests.get(url, allow_redirects=True)
open('TitanicSurvival.csv', 'wb').write(r.content)

with open('TitanicSurvival.csv') as csvfile:
    readCSV = csv.DictReader(csvfile)
    counter = 0
    ageSum = 0
    ageAvg = 0
    for row in readCSV:
      if row['age'] and float(row['age']) > 0:
        ageSum += int(float(row['age']))
        counter += 1
    if counter > 0:
      ageAvg = ageSum / counter
      print(ageAvg)
```
