Import data from GitHub

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import requests

url = 'https://github.com/mikekerklin/Machine-Learning-and-Data-Analysis/blob/main/diamonds.csv?raw=true'
df = pd.read_csv(url, index_col=0)
```

Describe numerical data

What are the unique values? Use the series method unique

```python
describe_unique = pd.Series(['carat'])
for carat in describe_unique:
    unique_values = df['carat'].unique().tolist()
    print(f"Unique values for {'carat'}: {unique_values}")

describe_unique = pd.Series(['cut'])
for cut in describe_unique:
    unique_values = df['cut'].unique().tolist()
    print(f"Unique values for {'cut'}: {unique_values}")

describe_unique = pd.Series(['color'])
for color in describe_unique:
    unique_values = df['color'].unique().tolist()
    print(f"Unique values for {'color'}: {unique_values}")

describe_unique = pd.Series(['clarity'])
for clarity in describe_unique:
    unique_values = df['clarity'].unique().tolist()
    print(f"Unique values for {'clarity'}: {unique_values}")

describe_unique = pd.Series(['depth'])
for depth in describe_unique:
    unique_values = df['depth'].unique().tolist()
    print(f"Unique values for {'depth'}: {unique_values}")

describe_unique = pd.Series(['table'])
for table in describe_unique:
    unique_values = df['table'].unique().tolist()
    print(f"Unique values for {'table'}: {unique_values}")

describe_unique = pd.Series(['price'])
for price in describe_unique:
    unique_values = df['price'].unique().tolist()
    print(f"Unique values for {'price'}: {unique_values}")

describe_unique = pd.Series(['x'])
for x in describe_unique:
    unique_values = df['x'].unique().tolist()
    print(f"Unique values for {'x'}: {unique_values}")

describe_unique = pd.Series(['y'])
for y in describe_unique:
    unique_values = df['y'].unique().tolist()
    print(f"Unique values for {'y'}: {unique_values}")

describe_unique = pd.Series(['z'])
for z in describe_unique:
    unique_values = df['z'].unique().tolist()
    print(f"Unique values for {'z'}: {unique_values}")
```
